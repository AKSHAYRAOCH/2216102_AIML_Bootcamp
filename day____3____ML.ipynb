```python
#class and object

class CO:
  def subn(self,a,b):
    return a-b
  def addn(self,a,b):
     return a+b

obj=CO()
print("addition of 2 numbers is",obj.addn(5,5))
print("sub of 2 numbers is ",obj.subn(10,5))
```

```python
class arithmetic:
  def addn(self,a,b):
    return a+b
  def subn(self,a,b):
    return a-b

  def mul(self,a,b):
    return a*b
  def div(self,a,b):
    return a/b
  def fldiv(self,a,b):
    return a//b
  def mod(self,a,b):
    return a%b
  def pow(self,a,b):
    return a**b

obj2=arithmetic()
print(obj2.addn(1,2))
print(obj2.subn(1,2))
print(obj2.mul(1,2))
print(obj2.div(1,2))
print(obj2.fldiv(1,2))
print(obj2.mod(1,2))
print(obj2.pow(1,2))
```

```python
#inheritance
#single
class Animal:
  def animal_sounds(self):
    return "makes sound"
class Cat(Animal):
  def cat_sound(self):
    return self.animal_sounds()+"MEOW"
  def __str__(self):
    return "CAT"
catobj=Cat()
print(catobj,catobj.cat_sound())
```

```python
#hierarchial
class Animal:
  def animal_sounds(self):
    return "makes sound"
class Cat(Animal):
  def cat_sound(self):
    return self.animal_sounds()+" MEOW"
  def __str__(self):
    return "CAT"
  
class Dog(Animal):
  def dog_sound(self):
    return self.animal_sounds()+" BOW"
  def __str__(self):
    return "DOG"
catobj=Cat()
print(catobj,catobj.cat_sound())
dogobj=Dog()
print(dogobj,dogobj.dog_sound())
```

```python
#multilevel inheritance
class Animal:
  def animal_sounds(self):
    return "makes sound"
class Cat(Animal):
  def cat_sound(self):
    return self.animal_sounds()+" MEOW"
class CatColor(Cat):
  def __str__(self):
    return "CAT"
  def color(self):
    return "color is black"
catcolor=CatColor()
print(catcolor,catcolor.cat_sound())
print(catcolor,catcolor.color())
```

```python
#exception handling
#2003
class INVALIDSTUDENTID(Exception):
  pass
stid=input("enter student id")
try:
  if stid[:4]!="2003":
    raise INVALIDSTUDENTID
except INVALIDSTUDENTID:
  print("error raised due to invalid details")
print("details verified")
```
